## 2.3 Epdemics

#### Exercise 2.9: Simulate an epidemic outbreak

```python
#Imports
import epidemics_helper
import json
import networkx as nx
from networkx.readwrite import json_graph
import numpy as np
from enum import Enum
import collections
import matplotlib.pyplot as plt
```

```python
#Load data
def read_json_file(filename):
    with open(filename) as f:
        js_graph = json.load(f)
    return json_graph.node_link_graph(js_graph)
```

```python
DATA_DIR = './../data/'

G = read_json_file(f'{DATA_DIR}nyc_augmented_network.json')
```

1. Lauch outbreak

```python
#Start simulation and lauch outbreak
beta = 10.0
gamma = 0.1

source = 23654
max_time = 100.0

def simulate_outbreak(G, beta, gamma, source, max_time):
    sir = epidemics_helper.SimulationSIR(G, beta=beta, gamma=gamma)
    sir.launch_epidemic(source=source, max_time=max_time)
    return sir
```

```python

sir = simulate_outbreak(G, beta, gamma, source, max_time)
```

2. Plot the evolution of the epidemic over time

```python
class State(Enum):
    susceptible = 0
    infectious = 1
    recovered = 2

def get_node_states(G, day):
    node_states = [sir.get_node_status(node, day) for node in sir.G.nodes()]
    return node_states

def get_all_node_states(G):
    days = np.arange(int(max_time))
    nodes = []
    for i in days:
        node_states = get_node_states(G, i)
        collection = collections.Counter(node_states)
        collection = dict(sorted(collection.items(),key = lambda i: i[0], reverse=False))
        nodes.append(collection)
    return days, nodes
        

days, nodes = get_all_node_states(G)
```

```python
width = 0.5
  
def get_state_percentage(sir, state):
    return np.array([node.get(state) if node.get(state) is not None else 0 for node in nodes]) / len(sir.G.nodes())

susceptible_vals = get_state_percentage(sir, 0)
bar1 = plt.bar(days, susceptible_vals, width, color = 'g')
  
infectious_vals = get_state_percentage(sir, 1)
bar2 = plt.bar(days, infectious_vals, width, color='r')
  
recovered_vals =  get_state_percentage(sir, 2)
bar3 = plt.bar(days, recovered_vals, width, color = 'b')

plt.xlabel("Days")
plt.ylabel('States')
plt.title("Outbreak data")
  
plt.xticks(days[::10] + 0.25, days[::10])
plt.legend( (bar1, bar2, bar3), ('susceptible', 'infectious', 'recovered') )
plt.show()
```

3. a) How long until 60% of the population is infected (at the same time)?

```python
np.argmax(infectious_vals >= 0.6)
```

3. b)  How long until 60% of the population is recovered (at the same time)?

```python
np.argmax(recovered_vals >= 0.6)
```

### 2.3.1 Stop the apocalypse!

#### Exercise 2.10: Strategy 1

```python
from random import sample, choice
edges_budget = 1000

G = read_json_file(f'{DATA_DIR}nyc_augmented_network.json')

def remove_edges(G, edges_budget):
    random_edges = sample(G.edges(), edges_budget)
    G.remove_edges_from(random_edges)
```

```python
def get_outbrak_stats(G, day):
    node_states = get_node_states(G, day)
    collection = collections.Counter(node_states)
    collection = dict(sorted(collection.items(),key = lambda i: i[0], reverse=False))
    return collection
```

```python
max_iter = 10
day = 30

def get_avg_stats(max_iter, day, edges_budget, remove_edge_mode=True):
    susceptible = []
    infectious = []
    recovered = []

    G = read_json_file(f'{DATA_DIR}nyc_augmented_network.json')
    source = choice(range(len(G.nodes())))
    if remove_edge_mode:
        remove_edges(G, edges_budget)

    for i in range(max_iter):
        sir = simulate_outbreak(G, beta, gamma, source, day)

        susceptible.append(np.sum(sir.status==0))
        infectious.append(np.sum(sir.status==1))
        recovered.append(np.sum(sir.status==2))
    return np.mean(susceptible), np.mean(infectious), np.mean(recovered)
```

```python
# without removing edges:
edges_budget = 1000
get_avg_stats(max_iter, day, edges_budget, remove_edge_mode=True)
```
